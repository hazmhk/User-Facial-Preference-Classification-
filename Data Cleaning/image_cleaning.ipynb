```python
import os
import cv2
import numpy as np
from PIL import Image
from retinaface import RetinaFace
import gc
```

```python

def is_real_life_image(image_path):
    # Placeholder for real-life vs. art detection logic
    # For now, it returns True by default
    return True

def process_images_in_batches(input_folder, output_folder, batch_size=50):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    image_files = [f for f in os.listdir(input_folder) if f.endswith(".jpg") or f.endswith(".png")]
    total_images = len(image_files)

    for start in range(0, total_images, batch_size):
        end = min(start + batch_size, total_images)
        batch_files = image_files[start:end]

        for filename in batch_files:
            image_path = os.path.join(input_folder, filename)
            image = cv2.imread(image_path)
            if image is None:
                print(f"Skipped {filename} - Unable to read image")
                continue

            detections = RetinaFace.detect_faces(image_path)
            if detections is None:
                print(f"Skipped {filename} - No faces detected")
                continue

            num_faces = len(detections)
            if num_faces == 1:
                for key, face in detections.items():
                    confidence = face['score']
                    if confidence > 0.95:  # Adjust the confidence threshold as needed
                        if is_real_life_image(image_path):
                            output_image_path = os.path.join(output_folder, filename)
                            cv2.imwrite(output_image_path, image)
                            print(f"Saved {filename} - One face detected with high confidence")
                        else:
                            print(f"Skipped {filename} - Not a real-life image")
                    else:
                        print(f"Skipped {filename} - Low confidence")
            else:
                print(f"Skipped {filename} - Faces detected: {num_faces}")

        # Clean up to free memory
        gc.collect()
        print(f"Processed batch {start//batch_size + 1} of {total_images//batch_size + 1}")

# Example usage
input_folder = '/content/drive/Othercomputers/My Laptop/results_bucket4'  # Adjust this path as needed
output_folder = '/content/drive/MyDrive/results4_filtered'  # Adjust this path as needed

# Process images to filter based on criteria
process_images_in_batches(input_folder, output_folder, batch_size=50)
```
